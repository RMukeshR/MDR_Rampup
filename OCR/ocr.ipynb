```python
from easyocr import Reader  # Correct

import pandas as pd

reader = Reader(['en'])  # Initialize EasyOCR with the language(s) you need
```

```python
import os
import cv2
folder_path = "Img_data/"

final_data=[]
for filename in os.listdir(folder_path):
    print(filename)
    current_file_path = os.path.join(folder_path, filename)
   
    # image = cv2.imread(current_file_path)

    # # Convert the image to grayscale
    # gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)



    results = reader.readtext(current_file_path)
    data = []
    data.append(filename)

    for detection in results:
        text = detection[1]
        data.append(text.split('\n'))
    data.append("//////")

    final_data.append(data)




```

```python
df = pd.DataFrame(final_data)

df.to_excel('/data1/home/mukeshram/MDR_Rampup/OCR/output_file_11.xlsx', index=False, header=True)
```
