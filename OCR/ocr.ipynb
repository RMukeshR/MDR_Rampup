```python
from paddleocr import PaddleOCR, draw_ocr
import os
import cv2
import matplotlib.pyplot as plt

ocr = PaddleOCR(use_angle_cls=True)
result = ocr.ocr("/data1/home/mukeshram/LLM/OCR/WhatsApp Image.jpeg")

# Extracting text boxes, texts, and scores from result
boxes = [line[0][0] for line in result]
txts = [line[1][0] for line in result]
scores = [line[1][1] for line in result]

# Specifying output path and font path
out_path = '/data1/home/mukeshram/MDR_Rampup/OCR/output'
font = './simfang.ttf'

def save_ocr(img_path, out_path, boxes, txts, scores, font):
    save_path = os.path.join(out_path, img_path.split('/')[-1] + 'output')

    image = cv2.imread(img_path)

    im_show = draw_ocr(image, boxes, txts, scores, font_path=font)

    cv2.imwrite(save_path, im_show)

    img = cv2.cvtColor(im_show, cv2.COLOR_BGR2RGB)
    plt.imshow(img)

# Call the function to save the OCR results
save_ocr("/data1/home/mukeshram/LLM/OCR/WhatsApp Image.jpeg", out_path, boxes, txts, scores, font)
```
